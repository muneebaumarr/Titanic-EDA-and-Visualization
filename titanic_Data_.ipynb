# **Titanic Dataset Analysis**

# **Introduction**


In this analysis, we explore the Titanic dataset to uncover insights into passenger demographics, survival rates, and factors influencing survival. The dataset contains information on passengers such as age, sex, fare, and class.

```python
#importing Datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#loading data
df= pd.read_csv('/content/tested.csv')

#display the first few rows of the dataset
df.head()
```

```python
#understanding The data
df.shape
```

```python
df.describe()
```

```python
df.head()
```

```python
#cleaning the data
df.drop(["PassengerId"] , axis=1 , inplace=True)
```

```python
df.drop(['Cabin'], axis=1, inplace=True)
```

```python
df.drop(['Name'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Sex'] = df['Sex'].map({'male':0, 'female':1})
```

```python
df.head()
```

```python
df['Embarked'] = df['Embarked'].map({'S':0, 'C':1, 'Q':2})
```

```python
df.head()
```

```python
df.drop(['Ticket'], axis=1, inplace=True)
```

```python
df.drop(['Fare'], axis=1, inplace=True)
```

```python
df.head()
```

```python
#filling null values

df.isnull().sum()
```

```python
df.isnull().sum()
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
#verfying there is no null value
df.isnull().sum()
```

```python
#data visulization
#Boxplot
#PairPlot
#Correlation Matrix
#Categorical Features
```

```python
df.head()
```

# **Box Plots**

```python
sns.boxplot(df['Age'])
```

```python
numerical_features = ['Age', 'SibSp', 'Parch']
categorical_features = ['Sex', 'Embarked']


for column in numerical_features:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x='Survived', y=column, data=df)
    plt.title(f'Box plot of {column} by Survival')
    plt.show()
```

```python
corr_matrix = df[['Age',  'Pclass', 'SibSp', 'Parch', 'Survived']].corr()

# Heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
pairplot_columns = ['Age', 'Pclass', 'Survived']
sns.pairplot(df[pairplot_columns], hue='Survived', markers=["o", "s"])
plt.show()
```

```python
categorical_features = ['Pclass', 'Sex', 'Embarked']
for column in categorical_features:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=column, hue='Survived', data=df)
    plt.title(f'Count plot of {column} by Survival')
    plt.show()
```

# **Survival Rate Analysis**

```python
#Trend Analysis
# Survival rate by passenger clas
pclass_survival_rate = df.groupby('Pclass')['Survived'].mean()
print(pclass_survival_rate)

```

```python
# Survival rate by gender

gender_survival_rate = df.groupby('Sex')["Survived"].mean()
print(gender_survival_rate)
```

```python
# Plot the survival rates
plt.figure(figsize=(10, 6))
sns.barplot(x=gender_survival_rate.index, y=gender_survival_rate)
plt.title('Survival Rate by Gender')
plt.ylabel('Survival Rate')
plt.xlabel('Gender')
plt.show()
```

```python
# Survival rate by embarkation point
embarked_survival = df.groupby('Embarked')['Survived'].mean()
print(embarked_survival)
```

```python

# Calculate the mean survival rate grouped by embarkation point
embarked_survival = df.groupby('Embarked')['Survived'].mean()

# Plot the survival rates
plt.figure(figsize=(10, 6))
sns.barplot(x=embarked_survival.index, y=embarked_survival)
plt.title('Survival Rate by Embarkation Point')
plt.ylabel('Survival Rate')
plt.xlabel('Embarkation Point')
plt.show()
```
